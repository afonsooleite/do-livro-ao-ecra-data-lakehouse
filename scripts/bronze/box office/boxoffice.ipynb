```python
from hdfs import InsecureClient

# 1) Conexão ao HDFS
client = InsecureClient('http://hdfs-nn:9870', user='hdfs')

# 2) Caminhos
from_path = "/home/jovyan/EDSTD/csv/boxoffice_data_2024.csv"   # caminho DENTRO do container
to_path = "/bronze/boxoffice_data_2024.csv"           # destino no HDFS

# 3) Criar pasta bronze se ainda não existir
client.makedirs("/bronze")

# 4) Fazer upload
client.upload(to_path, from_path, overwrite=True)

print("Upload completo!")
print(client.list("/bronze"))
 
```

```python
from pyspark.sql import SparkSession
from ydata_profiling import ProfileReport

# Criar sessão Spark
spark = SparkSession.builder \
    .appName("Profiling Box Office Data") \
    .master("local[*]") \
    .getOrCreate()

# Ler o CSV do HDFS
df_boxoffice = spark.read.option("header", "true").csv("hdfs://hdfs-nn:9000/bronze/boxoffice_data_2024.csv")

# Converter para Pandas (necessário para o profiling)
pdf_boxoffice = df_boxoffice.toPandas()

# Criar o relatório de Data Profiling
profile_boxoffice = ProfileReport(pdf_boxoffice, title="Box Office Data Profiling", explorative=True)

# Guardar o relatório em HTML
profile_boxoffice.to_file("/home/jovyan/profile_boxoffice.html")

print("Relatório gerado com sucesso!")
```
