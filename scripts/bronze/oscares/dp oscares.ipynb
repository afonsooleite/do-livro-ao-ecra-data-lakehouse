```python
from pyspark.sql import SparkSession
from ydata_profiling import ProfileReport
```

```python
# warehouse_location points to the default location for managed databases and tables
warehouse_location = 'hdfs://hdfs-nn:9000/warehouse'

spark = SparkSession \
    .builder \
    .master("local[2]") \
    .appName("Python Spark profiling example") \
    .config("spark.sql.warehouse.dir", warehouse_location) \
    .config("hive.metastore.uris", "thrift://hive-metastore:9083") \
    .enableHiveSupport() \
    .getOrCreate()
```

```python
from pyspark.sql import SparkSession

spark = SparkSession.builder \
    .appName("YData + Spark Teste") \
    .master("local[*]") \
    .getOrCreate()

spark
```

```python
df_full = spark.read.option("header", True).csv("./full_data (1).csv")
df_oscar = spark.read.option("header", True).csv("./the_oscar_award.csv")

```

```python
df_full.show(3)
df_oscar.show(3)
```

```python
pdf_full = df_full.toPandas()
pdf_oscar = df_oscar.toPandas()
```

```python
profile_full = ProfileReport(pdf_full, title="Relatório - Full Data", explorative=True)
profile_full.to_file("profile_full.html")
```

```python
profile_oscar = ProfileReport(pdf_oscar, title="Relatório - The Oscar Award", explorative=True)
profile_oscar.to_file("profile_oscar.html")
```
