```python
from hdfs import InsecureClient
import os


```

```python
client = InsecureClient("http://hdfs-nn:9870", user="anonymous")
```

```python
# Caminhos locais (ficheiros originais)
local_full = "./full_data (1).csv"   
local_oscar = "./the_oscar_award.csv"

# Caminhos no HDFS (fase Bronze)
hdfs_full = "/datasets/bronze/full_data/full_data.csv"
hdfs_oscar = "/datasets/bronze/the_oscar_award/the_oscar_award.csv"
```

```python
# Garantir que as pastas existem
client.makedirs("/datasets/bronze/full_data")
client.makedirs("/datasets/bronze/the_oscar_award")

# Fazer upload dos ficheiros originais
client.upload(hdfs_full, local_full, overwrite=True)
client.upload(hdfs_oscar, local_oscar, overwrite=True)


```
