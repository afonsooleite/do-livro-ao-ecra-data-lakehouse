```python
from pathlib import Path
from hdfs import InsecureClient
```

```python
client = InsecureClient("http://hdfs-nn:9870", user="anonymous")
```

```python
from_path = Path("../csvs/books.csv")
to_dir = "/datasets/bronze"
to_path = f"{to_dir}/books.csv"
```

```python
assert from_path.exists(), f"Ficheiro local n√£o encontrado: {from_path.resolve()}"
```

```python
client.makedirs(to_dir)
```

```python
client.delete(to_path, skip_trash=True)
```

```python
client.upload(
    hdfs_path=to_path,
    local_path=str(from_path),
    overwrite=True
)
```
