```python
from pathlib import Path
from hdfs import InsecureClient
```

```python
client = InsecureClient("http://hdfs-nn:9870", user="anonymous")
```

```python
from_path = Path("../csvs/Books_into_Movies_Atualizado_1920_2025.csv")

# Caminho DESTINO no HDFS (Bronze)
to_dir = "/datasets/bronze"
to_path = f"{to_dir}/Books_into_Movies_Atualizado_1920_2025.csv"
```

```python
assert from_path.exists(), f"Ficheiro local n√£o encontrado: {from_path.resolve()}"

client.makedirs(to_dir)
```

```python
client.upload(
    hdfs_path=to_path,
    local_path=str(from_path),
    overwrite=True
)
```
